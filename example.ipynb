### Necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
import glob
from PIL import Image
from lid_driven_cavity import *
```

### Parameters

```python
length = 1.0            # m  --  L                    
rho = 10                # kg*m-3 -- p                    
mu = 0.01               # Pa*s  -- mju                  
u = 1.0                 # m*s-1  --  U                    
n_p = 41                # number of points
iterations = 2000       # number of iterations    
dt = 0.001              # time step                    
iteration_poisson = 50  # poisson iteration
```

### Creating an object and making calculations

```python
obj = LidDrivenCavity(length = length, rho = rho, mu = mu, u = u, iter = iterations, dt = dt, n_p = n_p, iter_poisson = iteration_poisson, gif = True)
obj.print_reynolds()
obj.calculations()
u, v, p, X, Y = obj.result
plotting(u, v, p, X, Y )
```

```python
plt.figure()
plt.plot(Y, v[-round(obj.n_p/2), :])
plt.title('Centerline y-direction velocity')
plt.ylabel('v')
plt.xlabel('x/L')
plt.show()
```

```python
plt.figure()
plt.plot(Y, u[:, -round(obj.n_p/2)])
plt.title('Centerline x direction velocity')
plt.ylabel('u')
plt.xlabel('y/L')
plt.show()
```

```python
plt.figure()
plt.plot(u[:, -round(obj.n_p/2)], Y)
plt.show()
```
